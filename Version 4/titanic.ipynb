```python
train_df = pd.read_csv("/kaggle/input/titanic/train.csv")
# train_df = train_df.drop(columns=['Ticket', 'Cabin','PassengerId'])
train_df.columns
```

```python
test_df = pd.read_csv("/kaggle/input/titanic/test.csv")
# test_df = test_df.drop(columns=['Ticket', 'Cabin'])
# test_df
```

```python
train_df.iloc[:,1:]
```

```python
test_df.iloc[:,:]
```

```python
combine_df = pd.concat((train_df.iloc[:,2:], test_df.iloc[:,:]))
combine_df["Fam"] = combine_df['SibSp'] + combine_df['Parch']
# combine_df = combine_df.drop(['Name', 'SibSp', 'Parch'], axis=1)
combine_df
```

```python
combine_df.isnull().sum()
```

```python
age_temp = combine_df['Age'].to_numpy().reshape(-1, 1)
age_temp = StandardScaler().fit(age_temp).transform(age_temp)

combine_df['Age'] = age_temp
combine_df['Sex'] = combine_df['Sex'].replace("male", 0)
combine_df['Sex'] = combine_df['Sex'].replace("female", 1)
combine_df['Embarked'] = combine_df['Embarked'].replace("S", 0)
combine_df['Embarked'] = combine_df['Embarked'].replace("C", 1)
combine_df['Embarked'] = combine_df['Embarked'].replace("Q", 2)
combine_df = combine_df.fillna(0)

combine_df = pd.get_dummies(combine_df, dummy_na=True)
combine_df
```

```python
X_train = combine_df.iloc[:891,:]
X_test = combine_df.iloc[891:,:]
y_train = train_df.iloc[:,1]

X_train
```

```python
Random_modle = RandomForestClassifier()
Random_modle.fit(X_train,y_train)
pred_train=Random_modle.predict(X_train)
Random_modle.score(X_train, y_train)
```

```python
result = pd.DataFrame()
result['PassengerId'] = test_df['PassengerId']
result['Survived'] = Random_modle.predict(X_test)
result.to_csv('result.csv', index=False)
```
